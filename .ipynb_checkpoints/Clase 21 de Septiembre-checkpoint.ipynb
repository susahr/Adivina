```python
def walker ():
    global x,y
    eleccion = random.randint (0,3)
    if eleccion ==0:
        x+=1
    elif eleccion ==1:
        x-=1
    elif eleccion ==2:
        y+=1
    else:
        y-=1
    return [x,y]

x=0
y=0

Pos= np.zeros((2,10000))

for i in np.arange(0,len(Pos[0]),1):
    Pos [:,i] = walker()
    
plt.plot(Pos[0,:], Pos[1, :], 'bo-')
plt.plot(Pos[0,-1], Pos[1, -1], 'ro-')
```

```python
def walker():
    global x,y
    r = random.uniform(0,1)
    if r<0.01:
        x+= np.random.normal(0,200)
        y+= random.gauss(0,200)
    else:
        x+= np.random.normal(0,1)
        y+= random.gauss(0,1)
    return [x,y]
    
x=0
y=0

Pos = np.zeros((2,100))

for i in np.arange(0,len(Pos[0]),1):
    Pos [:,i] = walker()
    
plt.plot(Pos[0,:], Pos[1, :], 'bo-')
plt.plot(Pos[0,-1], Pos[1, -1], 'ro-')
```
