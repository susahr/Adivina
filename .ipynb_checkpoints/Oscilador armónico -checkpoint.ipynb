```python
t= np.linspace(0,50,100)
plt.figure(figsize =(7,4))
plt.plot(t,.5*np.cos(.5*t)+.1* np.sin(.5*t), '-', lw=1, ms=4,
         label = '$x(t)$')
plt.plot(t,.5*.1*np.cos(.5*t)-.5*.5*np.sin(.5*t), 'ro-', lw=1, ms=4,
         label = r'$\dot x(t)$')
plt.xlabel('$t$', fontsize=20)
plt.show()
```

```python
A=2
B=.1
W=.1
t= np.linspace(0,50,100)
plt.figure(figsize =(7,4))
plt.plot(t,A*np.cos(W*t)+B* np.sin(W*t), '-', lw=1, ms=4,
         label = '$x(t)$')
plt.plot(t,B*W*np.cos(W*t)-A*W*np.sin(W*t), 'ro-', lw=1, ms=4,
         label = r'$\dot x(t)$')
plt.xlabel('$t$', fontsize=20)
plt.legend(loc = 'best')
plt.show()
```
